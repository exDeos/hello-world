```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
filename = "3v3us_modtable.html"
with open(filename,'r',encoding='utf-8') as f:
    df = pd.read_html(f.read(),converters={'Class': str})[0]
df
```

```python
print("Top 5000 3v3 Records Spec Relative Frequency")
(df.groupby(["Class","Spec"]).size()/len(df)*100).sort_values(ascending=False)
```

```python
#print("Over 2800 Rating 3v3 Records Spec Relative Frequency")
rat2800 = (df[df.Rating>=2800].groupby(["Class","Spec"]).size()/len(df[df.Rating>=2800])*100).sort_values(ascending=False)
```
